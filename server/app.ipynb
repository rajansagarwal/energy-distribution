```python
import os
import pandas as pd
import markdown2

# Function to extract text from Markdown file
def extract_text_from_md(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        content = file.read()
        html_content = markdown2.markdown(content)  # Convert Markdown to HTML
        return html_content  # You can modify this for specific chunking logic

# Initialize a list to hold extracted data
data = []

# Path to the "notes" folder
notes_folder = 'notes'

# Walk through the directory tree
for root, dirs, files in os.walk(notes_folder):
    for file in files:
        if file.endswith('.md'):  # Check for Markdown files
            file_path = os.path.join(root, file)
            text = extract_text_from_md(file_path)
            data.append({'File': file, 'Text': text})

# Convert the list to a DataFrame
df = pd.DataFrame(data)

# Save DataFrame to CSV
csv_file = 'notes_data.csv'
df.to_csv(csv_file, index=False)

print(f"Data extracted and saved to '{csv_file}'")
```
